You arrive at the airport only to realize that you grabbed your North Pole Credentials instead of your passport. While these documents are extremely similar, North Pole Credentials aren't issued by a country and therefore aren't actually valid documentation for travel in most of the world.

It seems like you're not the only one having problems, though; a very long line has formed for the automatic passport scanners, and the delay could upset your travel itinerary.

Due to some questionable network security, you realize you might be able to solve both of these problems at the same time.

The automatic passport scanners are slow because they're having trouble detecting which passports have all required fields. The expected fields are as follows:
```

pid (Passport ID)
cid (Country ID)
```
Passport data is validated in batch files (your puzzle input). Each passport is represented as a sequence of key:value pairs separated by spaces or newlines. Passports are separated by blank lines.

The first passport is valid - all eight fields are present. The second passport is invalid - it is missing hgt (the Height field).

The third passport is interesting; the only missing field is cid, so it looks like data from North Pole Credentials, not a passport at all! Surely, nobody would mind if you made the system temporarily ignore missing cid fields. Treat this "passport" as valid.

The fourth passport is missing two fields, cid and byr. Missing cid is fine, but missing any other field is not, so this passport is invalid.

According to the above rules, your improved system would report 2 valid passports.

### Part One

### Part Two

The line is moving more quickly now, but you overhear airport security talking about how passports with invalid data are getting through. Better add some data validation, quick!

You can continue to ignore the cid field, but each other field has strict rules about what values are valid for automatic validation:
```
byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
pid (Passport ID) - a nine-digit number, including leading zeroes.
cid (Country ID) - ignored, missing or not.
```
Your job is to count the passports where all required fields are both present and valid according to the above rules. Here are some example values:

````python
# Approach:
    # Create funtion to convert file to list of dictionaires so that the fields can be easily accessed as k:v pairs
    # Create function for each of the more complex validity checks from part 2
    # Create function which increments a counter for pt1 and pt2 if meets the validity rules for each of those

import re

def create_passport_dicts(path:str) -> 'list(dict)':
    with open(path) as f:
        # splitting on '\n\n' splits file on blank rows,
        # then splitting again on spaces after converting remaining carriage returns to spaces
        passport_lists = [l.split() for l in [group.replace('\n', ' ') for group in f.read().split('\n\n')]]

        # converting to a list of dicts using the k:v pairs form each sublist
        passport_dicts = [dict(i.split(':') for i in passport) for passport in passport_lists]

    return passport_dicts

def has_valid_height(hgt:str) -> bool:
    '''
    Function which takes a height value as a string and returns a boolean
    - if in cm it is valid if height between 150 and 193
    - if in in it is valid if height between 59 and 76
    '''
    if 'cm' in hgt and 150 <= int(hgt.rstrip('cm')) <= 193:
        return True 
    elif 'in' in hgt and 59 <= int(hgt.rstrip('in')) <= 76:
        return True
    else:
        return False

def has_valid_hair_colour(hcl:str) -> bool:
    'Returns True if the input string is a hash symbol followed by 6 alpha numeric characters'
    return re.match(r'^#[a-zA-Z0-9]{6}$', hcl) is not None

def has_valid_eye_colour(ecl:str) -> bool:
    '''
    Returns True is the input string is in the list of valid eye colours:  
     - ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    '''
    return ecl in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']

def has_valid_passport_id(pid:str) -> bool:
    'Returns true if input is a nine-digit number, including leading zeroes'
    return pid.isdigit() and len(pid) == 9

def valid_passport_counter(path:str):
    '''
    - Functions checks each passport to see if if contains the correct fields (part one)
    - For all those that have the correct fields, checks them against the following validation checks:
        - byr (Birth Year) - four digits; at least 1920 and at most 2002.
        - iyr (Issue Year) - four digits; at least 2010 and at most 2020.
        - eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
        - hgt (Height) - a number followed by either cm or in:
            - If cm, the number must be at least 150 and at most 193.
            - If in, the number must be at least 59 and at most 76.
        - hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
        - ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
        - pid (Passport ID) - a nine-digit number, including leading zeroes.
    '''
    
    passport_dicts = create_passport_dicts(path)
    pt1_valid_passport_count = 0
    pt2_valid_passport_count = 0

    for passport in passport_dicts:
        passport_fields = list(passport)

        # Are all the required fields within the passport?
        if set(req_fields).issubset(set(passport_fields)):
            pt1_valid_passport_count += 1

            # Now check if each of those fields matches the specific field validations
            if  1920 <= int(passport['byr']) <= 2002   and \
                2010 <= int(passport['iyr']) <= 2020   and \
                2020 <= int(passport['eyr']) <= 2030   and \
                has_valid_height(passport['hgt'])      and \
                has_valid_hair_colour(passport['hcl']) and \
                has_valid_eye_colour(passport['ecl'])  and \
                has_valid_passport_id(passport['pid']):
                pt2_valid_passport_count += 1

    print(f'The number of passports with all the required fields is: {pt1_valid_passport_count}')  
    print(f'Out of these {pt2_valid_passport_count} meet all the required field level validation checks\n')  

    
sample_path  = 'Inputs\day_04_sample.txt'
sample_path2 = 'Inputs\day_04_sample2.txt'
main_path    = 'Inputs\day_04.txt'

valid_passport_counter(sample_path)
valid_passport_counter(sample_path2)
valid_passport_counter(main_path)
````

````python
# Improvements
# Alternative method which adds the validations into a 'validations dict' which can be compared against the passport dicts

import re

def create_passport_dicts(path:str) -> 'list(dict)':
    with open(path) as f:
        # splitting on '\n\n' splits file on blank rows,
        # then splitting again on spaces after converting remaining carriage returns to spaces
        passport_lists = [l.split() for l in [group.replace('\n', ' ') for group in f.read().split('\n\n')]]

        # converting to a list of dicts using the k:v pairs form each sublist
        passport_dicts = [dict(i.split(':') for i in passport) for passport in passport_lists]

    return passport_dicts

def has_valid_height(hgt:str) -> bool:
    '''
    Function which takes a height value as a string and returns a boolean
    - if in cm it is valid if height between 150 and 193
    - if in in it is valid if height between 59 and 76
    '''
    if 'cm' in hgt and 150 <= int(hgt.rstrip('cm')) <= 193:
        return True 
    elif 'in' in hgt and 59 <= int(hgt.rstrip('in')) <= 76:
        return True
    return False

def has_valid_hair_colour(hcl:str) -> bool:
    'Returns True if the input string is a hash symbol followed by 6 alpha numeric characters'
    return re.match(r'^#[a-zA-Z0-9]{6}$', hcl) is not None

def has_valid_eye_colour(ecl:str) -> bool:
    '''
    Returns True is the input string is in the list of valid eye colours:  
     - ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    '''
    return ecl in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']

def has_valid_passport_id(pid:str) -> bool:
    'Returns true if input is a nine-digit number, including leading zeroes'
    return pid.isdigit() and len(pid) == 9

def validate_fields(passport: dict) -> bool:
    validations = {
        'byr': lambda x: 1920 <= int(x) <= 2002,
        'iyr': lambda x: 2010 <= int(x) <= 2020,
        'eyr': lambda x: 2020 <= int(x) <= 2030,
        'hgt': has_valid_height,
        'hcl': has_valid_hair_colour,
        'ecl': has_valid_eye_colour,
        'pid': has_valid_passport_id
    }

    for field, validation in validations.items():
        if field not in passport or not validation(passport[field]):
            return False
    return True

def valid_passport_counter(path: str):
    passport_dicts = create_passport_dicts(path)
    pt1_valid_passport_count = 0
    pt2_valid_passport_count = 0

    for passport in passport_dicts:
        passport_fields = set(passport.keys())

        if set(req_fields).issubset(passport_fields):
            pt1_valid_passport_count += 1

            if validate_fields(passport):
                pt2_valid_passport_count += 1

    print(f'The number of passports with all the required fields is: {pt1_valid_passport_count}')
    print(f'Out of these, {pt2_valid_passport_count} meet all the required field level validation checks\n')


sample_path  = 'Inputs\day_04_sample.txt'
sample_path2 = 'Inputs\day_04_sample2.txt'
main_path    = 'Inputs\day_04.txt'

valid_passport_counter(sample_path)
valid_passport_counter(sample_path2)
valid_passport_counter(main_path)
````
