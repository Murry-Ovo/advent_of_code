You reach another volcanically active part of the cave. It would be nice if you could do some kind of thermal imaging so you could tell ahead of time which caves are too hot to safely enter.

Fortunately, the submarine seems to be equipped with a thermal camera! When you activate it, you are greeted with:

Congratulations on your purchase! To activate this infrared thermal imaging

Apparently, the Elves have never used this feature. To your surprise, you manage to find the manual; as you go to open it, page 1 falls out. It's a large sheet of transparent paper! The transparent paper is marked with random dots and includes instructions on how to fold it up (your puzzle input). For example:
```
6,10
0,14
9,10
0,3
10,4
4,11
6,0
6,12
4,1
0,13
10,12
3,4
3,0
8,4
1,10
2,14
8,10
9,0

fold along y=7
fold along x=5
```
The first section is a list of dots on the transparent paper. 0,0 represents the top-left coordinate. The first value, x, increases to the right. The second value, y, increases downward. So, the coordinate 3,0 is to the right of 0,0, and the coordinate 0,7 is below 0,0. The coordinates in this example form the following pattern, where # is a dot on the paper and . is an empty, unmarked position:
```
...#..#..#.
....#......
...........
#..........
...#....#.#
...........
...........
...........
...........
...........
.#....#.##.
....#......
......#...#
#..........
#.#........
```
Then, there is a list of fold instructions. Each instruction indicates a line on the transparent paper and wants you to fold the paper up (for horizontal y=... lines) or left (for vertical x=... lines). In this example, the first fold instruction is fold along y=7, which designates the line formed by all of the positions where y is 7 (marked here with -):
```
...#..#..#.
....#......
...........
#..........
...#....#.#
...........
...........
-----------
...........
...........
.#....#.##.
....#......
......#...#
#..........
#.#........
```
Because this is a horizontal line, fold the bottom half up. Some of the dots might end up overlapping after the fold is complete, but dots will never appear exactly on a fold line. The result of doing this fold looks like this:
```
#.##..#..#.
#...#......
......#...#
#...#......
.#.#..#.###
...........
...........
```
Now, only 17 dots are visible.

Notice, for example, the two dots in the bottom left corner before the transparent paper is folded; after the fold is complete, those dots appear in the top left corner (at 0,0 and 0,1). Because the paper is transparent, the dot just below them in the result (at 0,3) remains visible, as it can be seen through the transparent paper.

Also notice that some dots can end up overlapping; in this case, the dots merge together and become a single dot.

The second fold instruction is fold along x=5, which indicates this line:
```
#.##.|#..#.
#...#|.....
.....|#...#
#...#|.....
.#.#.|#.###
.....|.....
.....|.....
```
Because this is a vertical line, fold left:
```
#####
#...#
#...#
#...#
#####
.....
.....
```
The instructions made a square!

The transparent paper is pretty big, so for now, focus on just completing the first fold. After the first fold in the example above, 17 dots are visible - dots that end up overlapping after the fold is completed count as a single dot.

````python
import numpy as np
from matplotlib import pyplot as plt
# import itertools

input_file_list = []
sample_input_file_list = []
sample2_input_file_list = []
sample3_input_file_list = []

day = '13'

with open(f'Inputs\\day_{day}.txt', 'r') as input_file: 
    for line in input_file.readlines():
        input_file_list.append(line.rstrip('\n'))
with open(f'C:Inputs\\day_{day}_sample.txt', 'r') as input_file: 
    for line in input_file.readlines():
        sample_input_file_list.append(line.rstrip('\n'))
with open(f'C:Inputs\\day_{day}_sample2.txt', 'r') as input_file: 
    for line in input_file.readlines():
        sample2_input_file_list.append(line.rstrip('\n'))        
with open(f'C:Inputs\\day_{day}_sample3.txt', 'r') as input_file: 
    for line in input_file.readlines():
        sample3_input_file_list.append(line.rstrip('\n'))        
                     
# part one sample answer - 17 dots visible        
````

````python
# takes the inputs and splits it into the list of dot positions and the instructions
def get_dots_and_instrucs(raw_input: list) -> tuple:
    # turning instructions into (axis,int) format, e.g 'fold along y=7' will be (y, 7) -> (str, int)
    raw_instructions = raw_input[raw_input.index('')+1:]
    instructions = []
    for instruction in raw_instructions:
        instructions.append((instruction[11], # the x/y fold instruction
                             int(instruction.split('=')[1]))) # the the fold index
    # turning dots into a list of lists to make it easier to plot on a chart later 
    raw_dots = raw_input[:raw_input.index('')]
    dots = []
    for dot in raw_dots:
        x,y = dot.split(',')
        dots.append([int(x), int(y)])   
    return dots, instructions

# function to take the list of x,7 coords and plot them on a scatter plot
# and plot the fold line on the scatter plot 
def map_dots_and_fold(dots
                      ,fold_line
                      ,title
                      ,figure_size
                      ,dot_type):    
    # removes a warning bug present in numpy and matplot lib
    import warnings
    warnings.simplefilter(action='ignore', category=FutureWarning)
    
    plt.figure(figsize=figure_size)
    plt.title(title)
    
    # plots the dots on a scatter plot graph
    data = np.array(dots)
    x, y = data.T
    plt.scatter(x,y, c='#0A9928', marker=dot_type, s=168)  
    
    # plots the fold line on the x or y axis
    if fold_line[0] == 'x':
        plt.axvline(x=fold_line[1], color='#767575', linestyle='-')
    else:
        plt.axhline(y=fold_line[1], color='#767575', linestyle='-')
    
    # invert the y axis
    plt.gca().invert_yaxis()
    
    # Removing ticks and labels
    plt.xticks([]), plt.yticks([])
    plt.show()
    
# function to take the list of x,7 coords and plot them on a scatter plot
def map_dots(dots
             ,title
             ,figure_size
             ,dot_type
             ,axis_range=None):    
    # removes a warning bug present in numpy and matplotlib
    import warnings
    warnings.simplefilter(action='ignore', category=FutureWarning)
    
    plt.figure(figsize=figure_size)
    plt.title(title)
    
    # plots the dots on a scatter plot graph
    data = np.array(dots)
    x, y = data.T
    plt.scatter(x,y, c='#0A9928', marker=dot_type, s=168)   
       
    if axis_range == None:
        pass
    else:
        plt.xlim(axis_range[0]-1, axis_range[1]+1)
        plt.ylim(axis_range[2]-1, axis_range[3]+1)
    
    # invert the y axis
    plt.gca().invert_yaxis()
    
    # Removing ticks and labels
    plt.xticks([]), plt.yticks([])
    plt.show()    

# iterate through one fold instruction 
def process_fold_instruction(dots
                             ,instruction
                             ,fold_number
                             ,figure_size
                             ,dot_type):          
    x_vals = [coord[0] for coord in dots]
    y_vals = [coord[1] for coord in dots]
    
    max_x, min_x = max(x_vals), min(x_vals)
    max_y, min_y = max(y_vals), min(y_vals)
    
    fold_axis, fold_idx = instruction
    new_dots = []

    # for each dot if the x/y coord is less than the x/y fold line then add to new_dots
    # other wise invert the x/y coord using the following logic x - (x - fold_index)*2
    # if the x/y coord is equal to the fold line then do nothing
    if fold_axis == 'y':
        for dot in dots:
            x,y = dot
            # if the y value is lower than the fold line index then add to new_dots list (if not already present)
            if y < fold_idx:
                if dot not in new_dots:
                    new_dots.append(dot)
            # otherwise invert the y coord and append to new_dots (if not already present)
            elif y > fold_idx:
                new_dot = [x, y - ((y - fold_idx)*2)]
                if new_dot not in new_dots:
                    new_dots.append(new_dot)           
    elif fold_axis == 'x':
        for dot in dots:
            x,y = dot
            # if the y value is lower than the fold line index then add to new_dots list (if not already present)
            if x < fold_idx:
                if dot not in new_dots:
                    new_dots.append(dot)
            # otherwise invert the y coord and append to new_dots (if not already present)
            elif x > fold_idx:
                new_dot = [x - ((x - fold_idx)*2), y]
                if new_dot not in new_dots:
                    new_dots.append(new_dot)            
    else: print(f'Unkown fold axis {fold_axis}')  

    # plot the starting array with the fold line
    map_dots_and_fold(dots
                      ,instruction
                      ,title=f'Fold Instruction Number {fold_number}'
                      ,figure_size=figure_size
                      ,dot_type=dot_type)
    
    # plot the final state after the fold
    map_dots(new_dots
             ,title=f'Fold {fold_number} Outcome'
             ,figure_size=figure_size
             ,dot_type=dot_type
             ,axis_range = (min_x, max_x, min_y, max_y))
    return new_dots

# iterate through all fold instructions
def process_all_folds(dots
                      ,instructions
                      ,figure_size
                      ,dot_type): 
    dots_after_folds = []
    for inst_idx, instruction in enumerate(instructions):
        print(f'Fold Instruction: {inst_idx+1}\n- Fold Along The {instruction[0]} axis at index {instruction[1]}')
        new_dots = process_fold_instruction(dots
                                            ,instruction
                                            ,fold_number=inst_idx+1
                                            ,figure_size=figure_size
                                            ,dot_type=dot_type)
        
        dots_after_folds.append(len(new_dots))
        dots = new_dots  
    return dots_after_folds
````

````python
raw_input = input_file_list
# raw_input = sample_input_file_list

dots, instructions = get_dots_and_instrucs(raw_input)

dots_after_folds = process_all_folds(dots
                                     ,instructions
                                     ,figure_size=(9,3)
                                     ,dot_type='s')

for idx, fold_dot_count in enumerate(dots_after_folds):
    print(f'Dots After Fold Number {idx+1} = {fold_dot_count}')
````
