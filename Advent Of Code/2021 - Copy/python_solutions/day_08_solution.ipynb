```python
raw_input = sample_input_file_list
# raw_input = input_file_list

# converting the raw input into 2 lists, the first being the signal_patterns and second being the output_values
def convert_input(raw_input):
    print('-- converting raw input into signal_patterns and four digit output_values lists')
    signal_patterns, output_values = [], []
    for line in raw_input:
        # splitting each line based on the pipe delimiter then appending to the relevant list
        line_element = line.split("|")
        signal_patterns.append(line_element[0].strip())
        output_values.append(line_element[1].strip())

    return signal_patterns, output_values

# function that takes a string and retiurns the digit it represents
def assign_unique_numbers(input_string):
    if len(input_string) == 2:
        digit = '1'
    elif len(input_string) == 3:
        digit = '7'
    elif len(input_string) == 4:
        digit = '4'
    elif len(input_string) ==7:
        digit = '8'
    else:
        digit = ''
    return digit
    
# function takes the output values and returns a list of all the digits which can be dervied
#  due to them having a unique string length 
def get_unique_digits_from_output_values(output_values):
    unique_digits_in_output_values = []
    for line in output_values:
        line_split = line.split(' ')
        for coded_digit in line_split:
            digit = assign_unique_numbers(coded_digit)
            if digit != '':
                unique_digits_in_output_values.append(digit)
            else:
                pass
    print(f'there are {len(unique_digits_in_output_values)} unique digits within the output values')        
    return unique_digits_in_output_values 

signal_patterns, output_values = convert_input(raw_input)
unique_digits_in_output_values = get_unique_digits_from_output_values(output_values)
```

```python
# converting the raw input into 2 lists of lists
# the first being the signal_patterns and second being the output_values
def convert_input_v2(raw_input):
    print('-- converting raw input into signal_patterns and four digit output_values lists')
    all_signal_patterns, all_output_values = [], []
    for line in raw_input:
        # splitting each line based on the pipe delimiter, 
        signal_patterns, output_values = line.split("|")
        
        # splitting again based on spaces to get each decoded dig
        signal_patterns = signal_patterns.strip().split(' ')
        output_values = output_values.strip().split(' ')
        
        # sorting each decoded digit alphabetically (for easier matching later)
        signal_patterns = [''.join(sorted(x)) for x in signal_patterns]
        output_values = [''.join(sorted(x)) for x in output_values]
        
        # add the sorted lists to the main lists
        all_signal_patterns.append(signal_patterns)
        all_output_values.append(output_values)
    return all_signal_patterns, all_output_values

# function to identify the digit from the coded signal pattern, 
# logic is based on the volume of letters and whether the code shares some letters as an already known code
def decode_all_signal_patterns(signal_patterns):
    print('-- converting signal_patterns into a dict of the digit each pattern represents')
    '''
    Calculate the segments in this order:
    1, 4, 7, 8 - work out based on number of segments  
    3 - this will have 5 segments (len() == 5), but also will contain the same segments as 1  
    9 - has 6 segments and it shares the same segments as 4 and 3  
    0 - has 6 segments and it shares the same segments as 7  
    6 - has 6 segments (is the final 6 segment digit)   
    5 - has 5 segments and it shares the same segments as 6  
    2 - is the final 5 digit to decipher 
    '''
    # create empty dictionary to store all derived data
    all_sig_pats_dict = {}
    # iterate over each line in the signal patterns to identify which digits each pattern represents
    for index, signal_pattern in enumerate(signal_patterns):   
        sig_pat_dict = {} 
        known_digits = []           
        # identifying numbers 1, 4, 7 and 8 based on unique string lengths
        for unknown_digit in signal_pattern:
            if len(unknown_digit) == 2:
                sig_pat_dict['1'] = unknown_digit
                known_digits.append(unknown_digit)
            if len(unknown_digit) == 4:
                sig_pat_dict['4'] = unknown_digit
                known_digits.append(unknown_digit)
            if len(unknown_digit) == 3:
                sig_pat_dict['7'] = unknown_digit 
                known_digits.append(unknown_digit)
            if len(unknown_digit) == 7:
                sig_pat_dict['8'] = unknown_digit
                known_digits.append(unknown_digit)
                
        # identify digit 3 based on if it has 5 segments and contains all of the segments from 1
        for unknown_digit in signal_pattern:
            if unknown_digit in known_digits or len(unknown_digit) != 5:
                pass
            else:    
                matching_digit = sig_pat_dict['1']
                digit_counter = 0
                for letter in unknown_digit:
                    if letter in matching_digit:
                        digit_counter += 1
                if digit_counter == len(matching_digit):
                    sig_pat_dict['3'] = unknown_digit
                    known_digits.append(unknown_digit)       

        # identify digit 9 based on if it has 6 segments and contains all of the segments from 3
        for unknown_digit in signal_pattern:
            if unknown_digit in known_digits or len(unknown_digit) != 6:
                pass
            else:
                matching_digit = sig_pat_dict['3']
                digit_counter = 0
                for letter in unknown_digit:
                    if letter in matching_digit:
                        digit_counter += 1
                if digit_counter == len(matching_digit): 
                    sig_pat_dict['9'] = unknown_digit
                    known_digits.append(unknown_digit)           

        # identify digit 0 based on if it has 6 segments and contains all of the segments from 7
        for unknown_digit in signal_pattern:
            if unknown_digit in known_digits or len(unknown_digit) != 6:
                pass
            else:
                matching_digit = sig_pat_dict['7']
                digit_counter = 0
                for letter in unknown_digit:
                    if letter in matching_digit:
                        digit_counter += 1
                if digit_counter == len(matching_digit): 
                    sig_pat_dict['0'] = unknown_digit
                    known_digits.append(unknown_digit)

        # identify digit 6 based on if it has 6 segments (should be the final unknown digit with 6 segments)
        for unknown_digit in signal_pattern:
                if unknown_digit in known_digits or len(unknown_digit) != 6:
                    pass
                else:
                    sig_pat_dict['6'] = unknown_digit
                    known_digits.append(unknown_digit)           

        # identify digit 5 based on if it has 5 segments and all segments are contained within digit 6
        for unknown_digit in signal_pattern:
            if unknown_digit in known_digits or len(unknown_digit) != 5:
                pass
            else:                
                matching_digit = sig_pat_dict['6']
                digit_counter = 0
                for letter in unknown_digit:
                    if letter in matching_digit:
                        digit_counter += 1
                if digit_counter == len(unknown_digit):
                    sig_pat_dict['5'] = unknown_digit
                    known_digits.append(unknown_digit) 

        # identify digit 2 based on if it is the last digit to be identified
        for unknown_digit in signal_pattern:
            if unknown_digit in known_digits or len(unknown_digit) != 5:
                pass
            else:
                sig_pat_dict['2'] = unknown_digit
                known_digits.append(unknown_digit)
        all_sig_pats_dict[index] = sig_pat_dict 
    return all_sig_pats_dict    

# zipping together the list of output vals and decoded sig pat dicts then iterating
def decode_output_values(output_values, decoded_signal_patterns):
    print('-- converting the output values into a list of decoded output values using the decoded signal pattern dict')
    decoded_output_values = []
    # iterate over both output_values, decoded_signal_patterns at the same time
    for ovs, dsps in zip(output_values, decoded_signal_patterns.items()):
        dsps_dict = dsps[1]
        decoded_output_value = ''
        for unknown_ov in ovs:
            for key, value in dsps_dict.items():
                if value == unknown_ov:
                    decoded_output_value += key
                    break
        decoded_output_values.append(int(decoded_output_value))
    print(f'\nThe sum of all decoded output values is: {sum(decoded_output_values_list)}')
    return decoded_output_values    
    
# raw_input = sample_input_file_list
raw_input = input_file_list    

signal_patterns, output_values = convert_input_v2(raw_input)
decoded_signal_patterns = decode_all_signal_patterns(signal_patterns)
decoded_output_values_list = decode_output_values(output_values, decoded_signal_patterns)
```
