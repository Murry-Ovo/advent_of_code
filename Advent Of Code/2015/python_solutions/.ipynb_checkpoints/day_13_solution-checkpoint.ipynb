### Inputs and Imports

````python
from itertools import permutations

input_file_list = []
sample_input_file_list = []

day = '13'

with open(f'Inputs\\day_{day}.txt', 'r') as input_file: 
    for line in input_file.readlines():
        input_file_list.append(line.rstrip('\n'))
with open(f'C:Inputs\\day_{day}_sample.txt', 'r') as input_file: 
    for line in input_file.readlines():
        sample_input_file_list.append(line.rstrip('\n'))        

########################        
# Part One Sample Answer: 330 (David-Alice-Bob-Carol)
########################
# Part Two Sample Answer:
########################
````

### Part One

In years past, the holiday feast with your family hasn't gone so well. Not everyone gets along! This year, you resolve, will be different. You're going to find the optimal seating arrangement and avoid all those awkward conversations.

You start by writing up a list of everyone invited and the amount their happiness would increase or decrease if they were to find themselves sitting next to each other person. You have a circular table that will be just big enough to fit everyone comfortably, and so each person will have exactly two neighbors.

For example, suppose you have only four attendees planned, and you calculate their potential happiness as follows:

`Alice would gain 54 happiness units by sitting next to Bob.
Alice would lose 79 happiness units by sitting next to Carol.
Alice would lose 2 happiness units by sitting next to David.
Bob would gain 83 happiness units by sitting next to Alice.
Bob would lose 7 happiness units by sitting next to Carol.
Bob would lose 63 happiness units by sitting next to David.
Carol would lose 62 happiness units by sitting next to Alice.
Carol would gain 60 happiness units by sitting next to Bob.
Carol would gain 55 happiness units by sitting next to David.
David would gain 46 happiness units by sitting next to Alice.
David would lose 7 happiness units by sitting next to Bob.
David would gain 41 happiness units by sitting next to Carol.`

Then, if you seat Alice next to David, Alice would lose 2 happiness units (because David talks so much), but David would gain 46 happiness units (because Alice is such a good listener), for a total change of 44.

If you continue around the table, you could then seat Bob next to Alice (Bob gains 83, Alice gains 54). Finally, seat Carol, who sits next to Bob (Carol gains 60, Bob loses 7) and David (Carol gains 55, David gains 41). The arrangement looks like this:

```
     +41 +46
+55   David    -2
Carol       Alice
+60    Bob    +54
     -7  +83
```
After trying every other seating arrangement in this hypothetical scenario, you find that this one is the most optimal, with a total change in happiness of 330.

What is the total change in happiness for the optimal seating arrangement of the actual guest list?

````python
raw_input = input_file_list
# raw_input = sample_input_file_list

# take the raw input and outputs:
    # a distinct list of people to sit 
    # a dict containing all the pairings and the happiness units 
def parse_input(raw_input):
    pair_units_dict = {}
    people_list = []
    for i in raw_input:
        split = i.replace('.', '').split(' ') # removing the '.' and spoitting based on space   
        pair  = f'{split[0]}-{split[10]}' # the person pairing variable
        units = -int(split[3]) if split[2] != 'gain' else int(split[3]) # turning the units to negative if not gain
        pair_units_dict[pair] = units
        people_list.append(split[0]) #adding the first person to the list of people
    
    people_list = list(set(people_list)) # removing dupes from people list
    return pair_units_dict, people_list 

# takes the list of people and pairing dict and outputs:
# a list of all possible seating plans with the total happiness
def calc_splan_happy(pair_units_dict, people_list):   
    # create list of all possible combinations of guests
    s_plans = list(permutations(people_list))

    # iterate through all possible seating plans, work out the total happiness score and add to a dict
    s_plans_list = []
    for s_plan in s_plans:
        splan_happy = 0
        for idx, pers in enumerate(s_plan):
            pair     = f'{pers}-{s_plan[idx-1]}' # pairs this person with the one at position cur_index -1 
            rev_pair = f'{s_plan[idx-1]}-{pers}' # reversal of the above to esure we get the happy score both ways
            happy = pair_units_dict[pair] + pair_units_dict[rev_pair] # adding the happy scores of both pairing and reverse pairing
            splan_happy += happy
        # create a dict entry of the seating plan as the key and total happy as value
        s_plans_list.append(['-'.join(s_plan),splan_happy])

    return s_plans_list    

# caluclate the seating plan with the happiest guest order
def calc_happiest_plan(s_plans_list):
    # identify the maximum happy value
    all_the_happy = []
    for i in s_plans_list: all_the_happy.append(i[1])
    max_happy = max(all_the_happy)
    
    possible_plans = [x[0] for x in s_plans_list if x[1] == max_happy]
    print(f'The total change in happiness for the optimal seating plan is: {max_happy}')
    print('The possible seating plans with this happy value are: \n')
    for i in possible_plans: print(i)

pair_units_dict, people_list = parse_input(raw_input)
s_plans_list = calc_splan_happy(pair_units_dict, people_list)
calc_happiest_plan(s_plans_list)
````

### Part Two

In all the commotion, you realize that you forgot to seat yourself. At this point, you're pretty apathetic toward the whole thing, and your happiness wouldn't really go up or down regardless of who you sit next to. You assume everyone else would be just as ambivalent about sitting next to you, too.

So, add yourself to the list, and give all happiness relationships that involve you a score of 0.

What is the total change in happiness for the optimal seating arrangement that actually includes yourself?

````python
raw_input = input_file_list
# raw_input = sample_input_file_list

def add_me_to_list(people_list, pair_units_dict):
    # add a pairing of 'Me-Guest' and 'Guest-Me' to pair_units_dict
    for person in people_list:
        pair, rev_pair = f'Me-{person}', f'{person}-Me'
        pair_units_dict[pair] = 0
        pair_units_dict[rev_pair] = 0
        
    # add me to the list of people
    people_list.append('Me')
    
    return pair_units_dict, people_list

pair_units_dict, people_list = parse_input(raw_input)
pair_units_dict, people_list = add_me_to_list(people_list, pair_units_dict)
s_plans_list = calc_splan_happy(pair_units_dict, people_list)
calc_happiest_plan(s_plans_list)
````
