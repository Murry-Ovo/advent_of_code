# Day 13: Packet Scanners

You need to cross a vast firewall. The firewall consists of several layers, each with a security scanner that moves back and forth across the layer. To succeed, you must not be detected by a scanner.

By studying the firewall briefly, you are able to record (in your puzzle input) the depth of each layer and the range of the scanning area for the scanner within it, written as depth: range. Each layer has a thickness of exactly 1. A layer at depth 0 begins immediately inside the firewall; a layer at depth 1 would start immediately after that.

For example, suppose you've recorded the following:
```
0: 3
1: 2
4: 4
6: 4
```
This means that there is a layer immediately inside the firewall (with range 3), a second layer immediately after that (with range 2), a third layer which begins at depth 4 (with range 4), and a fourth layer which begins at depth 6 (also with range 4). Visually, it might look like this:

```

Within each layer, a security scanner moves back and forth within its range. Each security scanner starts at the top and moves down until it reaches the bottom, then moves up until it reaches the top, and repeats. A security scanner takes one picosecond to move one step. ````python
### Dev for part two

class Layer:
    # a class to represent an individual layer of the firewall with all it's relevant attributes
    def __init__(self, depth_, range_):
        self.depth_          = depth_
        self.scanner_range_  = range_
        self.severity_       = self.depth_ * self.scanner_range_
        self.scanner_loc_    = 0
        self.scanner_pattern = list(range(self.scanner_range_-1)) + list(range(self.scanner_range_-1, 0, -1))
            ## ^ the pattern the scanner will follow before repeating
            ## for example a scanner range of 3 will produce this pattern: [0,1,2,1]
        self.scanner_steps   = 0
        
    def __str__(self):
        ## allows for the printing of all the properties of the Layer class 
        return (f"depth            = {self.depth_}\n" 
                f"scanner range    = {self.scanner_range_}\n" 
                f"severity         = {self.severity_}\n" 
                f"scanner location = {self.scanner_loc_}\n"
                f"scanner pattern  = {self.scanner_pattern}\n"
                f"scanner steps    = {self.scanner_steps}\n")
    
    def move_scanner(self, picosecond_delay):
        '''
        - Function to move the scanner location to the next position within the scanner range
        - Based on the amount of steps already made it calculates the next position based on 
            - the scanner_pattern 
            - the picosecond_delay  
            - the number of scanner steps (starts incrementing regardless of the picosec-delay)
        '''
        pattern_element     = (self.scanner_steps + 1 + picosecond_delay) % len(self.scanner_pattern)
        self.scanner_steps += 1
        self.scanner_loc_   = self.scanner_pattern[pattern_element]
        
    def reset(self):
        '''
        Function to reset the scanner steps and the scanner location back to 0 for a layer
        '''
        self.scanner_steps = 0
        self.scanner_loc_  = 0
        
        
        
class Firewall:
    ## A class to represent the concept of the whole Firewall 
    def __init__(self, layers):
        
        # when instantiating 'Firewall' you need to pass a list of 'Layer'(s) in the format of the above class
        # This will be turned into the firewall_dict to use for the challenge 
        self.firewall_dict = {}
        
        for layer in layers:
            self.firewall_dict[layer.depth_] = layer
    
    def calculate_trip_severity(self, picosecond_delay):
        '''
        Function to calculate overall trip severity 
        
        '''
        picosec       = picosecond_delay
        current_depth = 0
        max_depth     = max(list(self.firewall_dict.keys()))
        trip_severity = 0
        
        for _ in range(max_depth+1):  
   
            if current_depth in self.firewall_dict:
                print(self.firewall_dict[current_depth])

                ## if the scanner location is at zero for the current dept (picosec = current_depth)
                ## then add the severity of the current_depth to the overall 'trip_severity'   
                if self.firewall_dict[current_depth].scanner_loc_ == 0:
                    trip_severity += self.firewall_dict[current_depth].severity_

            ## move the scanner to the next location for each depth
            for keys in self.firewall_dict.keys():
                self.firewall_dict[keys].move_scanner(picosecond_delay)

            ## increment the picoseconds
            current_depth += 1
        print(f'trip severity for picosecond_delay {picosecond_delay} = {trip_severity}\n-------')
        
        ## move the scanner to the next location for each depth
        for keys in self.firewall_dict.keys():
            self.firewall_dict[keys].reset()
                
        return trip_severity   

        
puzzle_input = sample
# puzzle_input = puz

layers = []        
for layer in puzzle_input:
    depth_, range_ = int(layer[0]), int(layer[1])
    layers.append(Layer(depth_, range_))   
    
picosecond_delay = 0
    
for _ in range(11):
    if picosecond_delay % 100000: print(f'checking picosecond delay: {picosecond_delay}')
        
    f = Firewall(layers)
    trip_severity = f.calculate_trip_severity(picosecond_delay)
    
    if picosecond_delay == 0:
        ## Part One Answer
        print(f'Running through the firewall at picosecond 0 has a trip severity of: {trip_severity}\n-------')
    
    if trip_severity == 0:
        ## Part Two Answer
        print(f'The first delay which produces a 0 trip severity is: {picosecond_delay}\n')
        break
    
    picosecond_delay += 1
    

````
